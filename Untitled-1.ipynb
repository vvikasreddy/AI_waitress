```python
import pyaudio
import audioop
import wave
import time

# Audio recording settings
FORMAT = pyaudio.paInt16  # 16-bit resolution
CHANNELS = 1              # Mono audio
RATE = 44100              # Sample rate (Hz)
CHUNK = 1024              # Buffer size
OUTPUT_FILE = "recorded_voice.wav"  # Output file name
SILENCE_THRESHOLD = 500   # RMS threshold for silence (adjust as needed)
SILENCE_DURATION = 2      # Seconds of silence before stopping

# Initialize PyAudio
audio = pyaudio.PyAudio()

# Open stream
stream = audio.open(format=FORMAT,
                    channels=CHANNELS,
                    rate=RATE,
                    input=True,
                    frames_per_buffer=CHUNK)

print("Recording started... Speak now!")

# Record audio until silence is detected
frames = []
silent_chunks = 0
chunks_per_second = RATE / CHUNK
silence_chunks_needed = int(SILENCE_DURATION * chunks_per_second)

while True:
    data = stream.read(CHUNK, exception_on_overflow=False)
    frames.append(data)
    
    # Calculate RMS (loudness)
    rms = audioop.rms(data, 2)  # 2 bytes per sample (16-bit)
    
    # Check if below silence threshold
    if rms < SILENCE_THRESHOLD:
        silent_chunks += 1
    else:
        silent_chunks = 0  # Reset if voice is detected
    
    # Stop if silence persists long enough
    if silent_chunks >= silence_chunks_needed:
        print("Silence detected. Stopping recording.")
        break

print("Recording finished.")

# Stop and close the stream
stream.stop_stream()
stream.close()
audio.terminate()

# Save the recorded audio to a WAV file
wf = wave.open(OUTPUT_FILE, 'wb')
wf.setnchannels(CHANNELS)
wf.setsampwidth(audio.get_sample_size(FORMAT))
wf.setframerate(RATE)
wf.writeframes(b''.join(frames))
wf.close()

print(f"Audio saved as {OUTPUT_FILE}")
```
